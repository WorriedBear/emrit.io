```python
common = pd.merge(fleet, denylist, how='inner', on=['hotspot_network_address'])
#common elements of fleet and denylist, along the hotspot_network_address column
```

```python
common = common.dropna(axis=1, how="all")
# common
# #finding all common values
```

```python
filtered_df = common[~common['lat'].isna()]
filtered_df
#common data values, where rows with NaN values removed
```
